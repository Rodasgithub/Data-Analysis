```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('retail.csv')
print(df)
```

```python
X = df[['age', 'income']].values
y = df['purchaseAmount'].values
```

```python
print("Retail")
model = LinearRegression()
model.fit(X, y)
```

### Age Vs Purchase Amount Relationship

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['purchaseAmount'])
plt.title('Age Vs Purchase Amount')
plt.xlabel('age')
plt.ylabel('purchase amount')
plt.xlim(16, 66)
plt.ylim(0, 1100)
plt.plot(X, model.predict(X), color='red')
plt.show()
```

### Income Vs Purchase Amount Relationship

```python
plt.figure(figsize=(12, 6))
plt.scatter(df['income'], df['purchaseAmount'])
plt.title('Income Vs Purchase Amount')
plt.xlabel('income')
plt.ylabel('purchase amount')
plt.xlim(19000, 100000)
plt.ylim(0, 1100)
plt.plot(X, model.predict(X), color='red')
plt.show()
```

### Slope

```python
print("Slope = " + str(model.coef_))
```

### Intercept

```python
print("Intercept = " + str(model.intercept_))
```

### Prediction

```python
print("A person aged 34 with income of 33,000 will have a purchase amount of "+str(model.predict([[34, 33000]])))
print("A person aged 57 with income of 25,780 will have a purchase amount of "+str(model.predict([[57, 25780]])))
print("A person aged 24 with income of 76, will have a purchase amount of "+str(model.predict([[57, 25780]])))
```
